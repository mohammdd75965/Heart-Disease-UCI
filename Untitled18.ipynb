```python
data=pd.read_csv('/content/drive/MyDrive/archive/fashion-mnist_train.csv')
```

```python
train_data=np.array(data,dtype='float32')
train_data
```

```python
test=pd.read_csv('/content/drive/MyDrive/archive/fashion-mnist_test.csv')
```

```python
test_data=np.array(test,dtype="float32")
```

```python
test_data
```

```python
x_train=train_data[:,1:]
x_train=x_train/255
```

```python
x_train
```

```python
y_train=train_data[:,0]/255
```

```python
x_test=test_data[:,1:]/255
y_test=test_data[:,0]/255
```

```python
image_rows = 28
image_cols = 28
batch_size = 4096
image_shape = (image_rows,image_cols,1) 
```

```python
x_train = x_train.reshape(x_train.shape[0],*image_shape)
x_test = x_test.reshape(x_test.shape[0],*image_shape)
import keras
```

```python
KerasModel = keras.models.Sequential([
        keras.layers.Conv2D(64,kernel_size=(3,3),activation='relu',input_shape=image_shape),
       
        keras.layers.MaxPool2D(4,4),
          
        keras.layers.Conv2D(32,kernel_size=(3,3),activation='relu'),
        keras.layers.MaxPool2D(2,2),
        keras.layers.Flatten(),    
        keras.layers.Dense(16,activation='relu') ,    
             
        keras.layers.Dropout(rate=0.5),            
        keras.layers.Dense(10,activation='softmax') ,    
        ])
```

```python
KerasModel.compile(optimizer ='adam',loss='sparse_categorical_crossentropy',metrics=['accuracy'])
```

```python
epochs = 10
ThisModel = KerasModel.fit(x_train,y_train, epochs=epochs,batch_size=batch_size,verbose=1)
```
